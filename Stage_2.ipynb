```python
import pandas as pd
```

```python
clean_data = pd.read_csv("cleaned_-data.csv")
```

```python
clean_data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.scatterplot(x='Weight', y='Grip strength', data=clean_data)
plt.title("Scatter plot of Grip strength vs weight")
plt.savefig("scatter__grip.png")
plt.show()
```

```python
sns.distplot(clean_data['Grip strength'] , kde= True, bins = 30, color = 'red')
plt.title("Distribution plot of Grip strength")
plt.savefig("distri__grip.png")
plt.show()
```
